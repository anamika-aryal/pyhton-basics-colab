```python
# importing librarires
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Note: seaborn has built in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
#setting a theme for plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x = 'total_bill', y = 'tip', col='time', hue ='smoker',style='smoker',size='size')
```

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
house_california = fetch_california_housing()

california_dataframe = pd.DataFrame(house_california.data,columns=house_california.feature_names)
california_dataframe['price']=house_california.target
```

```python
california_dataframe.head()
```

Distribution Plot

```python
sns.displot(california_dataframe['price'])
```

Correlation matrix


*   positive
*   negative



```python
correlation=california_dataframe.corr()
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'fontsize':8},cmap='Blues')
```
