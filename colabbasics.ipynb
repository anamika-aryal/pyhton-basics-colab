```python
# list -> [elements], can contain mutiple data type, but set dosn't allow it.
#list allows duplcate value
my_list=[1,True,3.78,'numbers']
print(my_list)
type(my_list)
```

```python
#add elements to the list:
my_list=[1,True,3.78,'numbers']
my_list.append(6)
print(my_list)
print(len(my_list))
```

```python
#add elements to the list:
my_list=[1,True,3.78,'numbers']
print(my_list)

del my_list[1]
print(len(my_list))
print(my_list)
```

```python
#join two list
list1=[1,2,3,4,5]
list2=[6,7,8,9,10]

list3=list1+list2
print(list3)
```

```python
#tuple
tuple1=(1,3.4,True,'AI')
print(tuple1)

my_list=[1,3,5,7,9]
print(my_list)
my_tuple=tuple(my_list)
print(my_tuple)
```
